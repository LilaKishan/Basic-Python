### 01) WAP to check given string is palindrome or not.

```python
s=input("enter String:")
if(s==(s[::-1])):
  print("palidrome")
else:
  print("not palidrome")
```

### 02) WAP to reverse the words in given string.

```python
s=input("enter String:")
y=s.split()

z=" ".join(y[::-1])
print(z)
```

### 03) WAP to remove ith character from given string

```python
s=input("enter String:")
i=int(input("enter ith index for remove:"))
print(s[:i:]+s[i+1::])
```

### 04) WAP to find length of String without using len function.

```python
s=input("enter String:")
count=0
for i in s:
  count+=1
print(f"string length is {count}")
```

### 05) WAP to print even length word in string.

### 06) WAP to count numbers of vowels in given string.

```python
y=input("enter String:")
s=y.lower()
count=s.count('a')+s.count('e')+s.count('i')+s.count('o')+s.count('u')
print(f"No. of vowels in string : {count}")
```

### 07) WAP to convert given array to string.

```python
arr = []
str1 = ""
length = int(input("Enter Length Of Charcter : "))
for i in range(length):
    character = input("Enter Character : ")
    arr.append(character)
str1 = str1.join(arr)
print(str1)
```

### 01) WAP to find out duplicate characters in given string.

```python
string = input("Enter a string ")
dictionary = {}
for i in string:
    if(string.count(i)>1 ):
        dictionary[i] = string.count(i)
print(dictionary)
```
