12) Compute distance between two points taking input from the user (Pythagorean Theorem).

```python
a1=int(input("enter point x1: "))
b1=int(input("enter point y1: "))
a2=int(input("enter point x2: "))
b2=int(input("enter point y2: "))
c=((((a2-a1)**2)+((b2-b1)**2))**0.5)
print("distance: ",c)
```
